
# DTIME(T(n)):
The DTIME(T(n)) is the set of languages decidable (by a deterministic TM) in time c*T(n) for some constant c.

Why is this classification not robust?
It is dependent on the computational model chosen, classes too small.

# Class P:
The set of languages decidable by a deterministic TM in polynomial time. Can be defined also as class P = U c>=1 DTIME(n^c).

# NP:
A language L⊆{0,1}∗ is in the class NP iﬀ there exist a polynomial p : N→N and a polynomial time TM M such that :

L = {x ∈{0,1}∗ |∃y ∈{0,1}^p(|x|).M(>x,y<) = 1}

- M is said to be the veriﬁer for L. 

- Any y ∈{0,1}^p(|x|) such that M(>x,y<) = 1 is said to be a certiﬁcate FOR X.

- Differently from P and EXP, the class NP does not have a natural counterpart as a class of functions (difficult to be defined).

-  If someone asks you an example of a problem which DOES NOT HAVE a candidate provable in polynomial time YOU DON'T KNOW IT, IF YOU KNEW IT YOU WOULD HAVE PROVED THAT EXP != NP, WHICH IS STILL UNKNOWN. 

# NDTM:
It has 2 transition functions, and at each step it choose non-deterministically one of them. 

There's a special state qaccept (not only halt(=reject) ant init; it means that it can happen that the TM halts(=reject), but does not accept!).

- We say the NDTM Accepts the input x ∈{0,1}∗ iﬀ THERE EXIST ONE among the many possible evolutions of the machine M when fed with x which makes it reaching qaccept. 

- Rejects the input x ∈{0,1}∗ iﬀ none of the aforementioned evolutions leads to qaccept.

# NDTM runtime:
We say that a NDTM M runs in time T : N→N iﬀ for every x ∈{0,1}∗ and FOR EVERY POSSIBLE nondeterministic evolution, M reaches either the qhalt or qaccept within c·T(|x|) steps, where c > 0.

# NDTIME(T(n)):
NDTIME(T(n)) is the set of languages decidable by a NDTM in time T.

# NP alternative definition:
By using the definition of NDTM we can define the class NP as:

NP = ∪c∈N  NDTIME(n^c)    for c>=1.

# pre-order and  REDUCTION

REDUCTION =  The language L is said to be polynomial-time reducible to another language H iﬀ there is a polytime computable function f : {0,1}∗ →{0,1}∗ such that x ∈L iﬀ f(x) ∈ H.  In this case, we write L≤p H (= THIS IS READ L IS POLYTIME REDUCIBLE TO H).

- Reduction is a PRE-ORDER RELATION: a binary relation which is transitive and reflexive. Why do we need it? to put in DIFFICULTY relation the languages. Indeed L ≤p H means that L is at most as difficult as H, since it is possible to convert it in the H Language and solve it like it. (THIS RELATION IS NOT ANTYSYMMETRIC (INDEED IF A<=B AND B<=A THEN is NOT true that A=B, anyway in that case I can say that they're equivalent in terms of complexity).

- reduction is a function, so if you want to prove the reduction between two languages you need to prove that a function with these properties exists:

  
     - f computable in polynomial time.
         
    - Domain and codomain are {0,1}*.
 
    - On input outside of L the output goes outside of H. If the input is a binary string belonging to L, then the output belongs to H. In this way we can prove that INSTANCES OF THE PROBLEM L CAN BE TRANSFORMED/REDUCED IN THE INSTANCES OF THE PROBLEM H, IN A FAITHFUL WAY. INDEED YOU DO A REDUCTION FROM L1 TO L2 ("L1 ≤p L2") IFF THE INSTANCES OF L1, let's call them x1, ARE TRANSFORMED INTO INSTANCES OF L2, let's call them x2, SUCH THAT x1 IS IN L1 IFF x2 IS IN L2. 


# class FP

What is it? Is exactly the same thing of the class P, but for "function problem". What are the function problem? Are those problems whose output is NOT NECESSARLY only 1/0. Remember this, because there are certain class of functions which can be converted, in a canonical way, from function to a language (so from a problem which takes a string as input and outputs a string, into a problem which simply outpus 0/1), for instance the functions (=problems) of optimization!! note that by optimization function we mean that the optimization problem itself is considered as a function, which taken a mathematical function (converted in string) outputs the variables in which there's the maximum (converted in string), and this holds:
- any Language which belongs to P can be transformed in a function which belongs to FP (since of course the Boolean functions which are in P are also in FP).
- also some kind of functions (=optim. problems) can be transformed in a language (because the problem becomes: "is the input above a certain threshold?"). **In general, however, it is not true that the language associated to a function in FP is also in P**.


# Class EXP and FEXP:

Are the same as P and FP, but instead of polynomial the bound is $2^{n^c}$

Example: the wedding plan /MIS problem.

Is there a class of algorithms which is something between EXP and P? YES! It is NP. Note that it doesn't exist FNP.<br>
Note that there are also algorithms which are not solvable in EXP time, for instance the halting problem which is undecidable.

**Note**: We proved that P is different from EXP, so it is strictly included. What about NP? Since we know that P is different from EXP, then NP must strictly include P OR strictly be included in EXP. We still don't know between the two options which is the correct one yet.

**Theorem** <br>  
P ⊆ EXP, FP ⊆ FEXP, these two inclusions are strict.



# NP-HARD:
L is NP-hard iff any other language in NP can be reduced to it.  ( note that to be reduced is more strong than simply being more difficult, indeed it means that the language L is a kind of "universal" language which can be used to solve other problems).


# NP-complete:
L is in NP-complete iff it is NP-HARD and it is in NP.


# k-CNF:
The disjunctions in a CNF are said to be clauses, and a kCNF is a CNF whose clauses contains AT MOST k ∈N literals. Examples: the two formulas X ∨¬X and X ∧(Y ∨¬Z) are 2CNFs, but not 1CNFs.

k low -> complexity of kSAT is lower.

- We need now to prove that EXP is different from P. We want to prove it by proving a stronger thing: the Hierarchy Theorem:
The Hierarchy Theorem informally says that given more time, a Turing machine can solve more problems.
More formally, the Hierarchy Theorem states that if f(n) and g(n) are time-constructable functions ad f(n)log(f(n)) = o(g(n)) then DTIME(f(n)) $\subset$ DTIME(g(n)) strictly.

We prove only a case of the HT theorem: ...  DTIME(n) $\subset$ DTIME(n^2) strictly.

# 5) P ⊆ NP ⊆ EXP

proof:

  1) P ⊆ NP. To prove it we must show that if L $\in$ P, then L can be written as follows: 
  
     L={x $\in$ {0,1}* | |∃ y  $\in$ {0,1}^p(|x|). M(>x,y<)=1} where p is a polynomial and M is poly-time computable.
     
     We can do it because we can simply choose p(n)=0 and M as a polytime deterministic machine deciding L (M exists because of the hypothesis that L is in P because we can use the function which solves the problem as a verifier, which ineed is taking as input the empty string+ the input x). This way indeed we obtain  L={x $\in$ {0,1}* | |∃ y  $\in$ {0,1}^0. M(>x,y<)=1} where y is the empty string, so it is equal (with a small encoding difference) to L={x $\in$ {0,1}* | M(>x<)=1}.
     
  2) NP ⊆ EXP. To prove it we must show that any NP problem can be solved in EXP time. To do that we simply say that any language L $\in$ NP can be DECIDED by trying all possible certificates until you find one for which the pair (input,certificate) satisfy the test of the verifier:
     
       input = x $\in$ {0,1}*
       
       for each y in {0,1}^p(|x|) do:
         
         - simulate M(x,y). M(x,y) output =b.
           
           if b=1 -> return 1
         
       return 0 
       
    We need to show that this has exponential time wrt the input. Well, the loop is done for a number of times which is equal to the cartinality of {0,1}^p(|x|) which is 2^p(|x|). Each iteration works in polynomial time because the TM M is poly-time wrt its input, which is itself polynomially bounded. We can call q(|x|) the polynomial time bound of each iteration. So the algorithm is bounded by 2^p(|x|) * q(|x|) = O( 2^p(|x|) * 2^{log(q(|x|))} ) = O( 2^ {p(|x|) + q(|x|)} = O(2^{p(|x|)+q(|x|)}) which is bounded by 2^{n^c} with c big enough, so it is exponential.

# 9) Cuk levain theorem:

Let SAT be the language of all satisﬁable CNF formulae and 3SAT be the language of all satisﬁable 3CNF formulae. Then,

1. SAT is NP-complete.

2. 3SAT is NP-complete.

Not a formal proof. Just an idea.

The proof is structured in this way:
 
First prove that SAT is NP-complete:

  1) show that SAT is in NP.

  2) completeness: any language L in NP can be reduced to SAT (L ≤p SAT).
     - define a poly-time transformation from string x to CNF formula phi, such that  phi is in SAT IFF x is in L.
     
Then show that 3SAT is NP-complete by simply showing this: SAT ≤p 3SAT & 3SAT is in NP.
